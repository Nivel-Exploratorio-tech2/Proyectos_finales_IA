```python
# Descargar los archivos (Asegúrate de que wget está disponible en tu entorno)
for archivo in archivos:
    file_id = archivo['file_id']
    nombre_archivo = archivo['nombre_archivo']

    comando = f"""
    wget --load-cookies /tmp/cookies.txt "https://docs.google.com/uc?export=download&confirm=$(wget --quiet --save-cookies /tmp/cookies.txt --keep-session-cookies --no-check-certificate 'https://docs.google.com/uc?export=download&id={file_id}' -O- | sed -rn 's/.*confirm=([0-9A-Za-z_]+).*/\\1\\n/p')&id={file_id}" -O {nombre_archivo} && rm -rf /tmp/cookies.txt
    """
    os.system(comando)

# Lista para almacenar los DataFrames filtrados
dataframes = []

# Filtrar por Manizales y agregar a la lista
for archivo in archivos:
    nombre_archivo = archivo['nombre_archivo']
    try:
        # Cargar el archivo CSV
        df = pd.read_csv(nombre_archivo)

        # Asegúrate de que la columna correcta exista (reemplaza 'ciudad' si es necesario)
        if 'Mercado' in df.columns:
            df_filtrado = df[df['Mercado'].str.contains('Manizales', case=False, na=False)]
            dataframes.append(df_filtrado)
        else:
            print(f"La columna 'Mercado' no se encuentra en {nombre_archivo}")
    except Exception as e:
        print(f"Error al cargar {nombre_archivo}: {e}")
```

```python
# Concatenar todos los DataFrames filtrados
df_concatenado = pd.concat(dataframes, ignore_index=True)
```

```python
print(df_concatenado.shape)  # Muestra el número de filas y columnas
df_concatenado  # Muestra las primeras filas del DataFrame
```

```python
# Contador de productos completados
total_completados = 0
```

```python
# Proceso de cálculo y completación
productos_faltantes = []  # Lista para almacenar los productos que serán añadidos
```

```python
# Contar el número de apariciones de cada producto
conteo_productos = df_concatenado['Producto'].value_counts()
conteo_productos
```

```python
# Identificar productos que aparecen menos de 12 veces
productos_bajos = conteo_productos[conteo_productos < 12].index
```

```python
print(productos_bajos)
```

```python
for producto in productos_bajos:
    # Calcular cuántas veces aparece el producto
    cantidad_actual = conteo_productos[producto]
    faltantes = 12 - cantidad_actual

    # Obtener el precio promedio del producto
    promedio_precio = df_concatenado[df_concatenado['Producto'] == producto]['Precio_kg'].mean()  # Cambia 'valor' al nombre de la columna que contiene los precios

    # Crear nuevas filas hasta completar 12 apariciones
    for _ in range(faltantes):
        nueva_fila = df_concatenado[df_concatenado['Producto'] == producto].iloc[0].copy()  # Copiar la fila existente
        nueva_fila['Precio_kg'] = promedio_precio  # Asignar el precio promedio
        productos_faltantes.append(nueva_fila)  # Agregar a la lista de productos faltantes

    total_completados += faltantes
```

```python
# Agregar las nuevas filas al DataFrame original
df_concatenado = pd.concat([df_concatenado, pd.DataFrame(productos_faltantes)], ignore_index=True)
```

```python
print(f"Total de productos completados: {total_completados}")
```
